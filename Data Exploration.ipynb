```python
import matplotlib.pyplot as plt
import rasterio

import os
os.chdir('/home/jovyan/EYODSC2024')
# Load the two TIFF files
pre_map = "Pre_Event_San_Juan.tif"
post_map = "Post_Event_San_Juan.tif"

# Open the TIFF files using rasterio
with rasterio.open(pre_map) as pre_src, rasterio.open(post_map) as post_src:
    
    # Get the coordinate boundaries
    pre_bound = pre_src.bounds
    post_bound = post_src.bounds

    print(pre_bound, post_bound)
    print(pre_src.crs, post_src.crs)
    
```

```python
# Open the TIFF files using rasterio
with rasterio.open(pre_map) as pre_src, rasterio.open(post_map) as post_src:

    #Read first grid
    i=31
    pre_data = pre_src.read(1, window=((1024*i, 1024*(i+1)), (1024*i, 1024*(i+1))))
    post_data = post_src.read(1, window=((1024*i, 1024*(i+1)), (1024*i, 1024*(i+1))))

        
    # Plot pre_data
    plt.subplot(1, 3, 1)
    plt.imshow(pre_data, cmap='gray')
    plt.title('Pre Data')

    # Plot post_data
    plt.subplot(1, 3, 2)
    plt.imshow(post_data, cmap='gray')
    plt.title('Post Data')

    # Plot the difference
    plt.subplot(1, 3, 3)
    plt.imshow(pre_data - post_data, cmap='gray')
    plt.title('Difference')

    # Adjust the layout
    plt.tight_layout()

    # Show the plots
    plt.show()
```
